Finally, the next few cells walk you through one form of analysis you might want to do on your collection of tweets. This one counts the number of positive and negative words in your tweet, with positive and negative words pulled from predetermined lists.

```python
# Create lists of positive and negative words.
posWords = open('positive-words.txt', 'r')
negWords = open('negative-words2.txt', 'r') # negative-words.txt will give error because of words spelled with more than one *
```

```python
posWordList = []
for line in posWords:
    #print line
    nextWord = line.strip()
    if len(nextWord) > 0 and not nextWord.startswith(";"):
        posWordList.append(nextWord)

posWordList
```

```python
negWordList = []
for line in negWords:
    #print line,
    nextWord = line.strip()
    if len(nextWord) > 0 and not nextWord.startswith(";"):
        negWordList.append(nextWord)

negWordList
```

```python
# For each tweet, look for each word.
# Keep a running tally of positive and negative words.
totalPos = 0
totalNeg = 0


for tweet in matchedTweets["statuses"]:
    tweetText = tweet["text"]
    tweetText = tweetText.encode('utf-8')
    tweetText = tweetText.strip()
    #print(tweetText)
    
    # Look for any positive words.
    for nextWord in posWordList:
        #print(nextWord)
        searchPattern = re.compile("\\b" + nextWord + "\\b")
        searchMatches = searchPattern.findall(tweetText.decode('utf-8'))
        #if len(searchMatches) > 0:
        #    print(tweetText, str(len(searchMatches)), nextWord)
        #print(searchMatches)
        totalPos = totalPos + len(searchMatches)
    
    # Look for any negative words.
    for nextWord in negWordList:
        #print(nextWord)
        searchPattern = re.compile('\\b' + nextWord + '\\b')
        searchMatches = searchPattern.findall(tweetText.decode('utf-8'))
        #if len(searchMatches) > 0:
        #    print(tweetText, str(len(searchMatches)), nextWord)
        #print(searchMatches)
        totalNeg = totalNeg + len(searchMatches)

        
print(totalPos)
print(totalNeg)
```
