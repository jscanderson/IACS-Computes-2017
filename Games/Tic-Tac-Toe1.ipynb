```python
def draw_board(p1, p2):
    '''
        Function to draw the board with correct symbols
        in each square. Input parameters are:
            p1 - list of squares player1 controls
            p2 - list of squares player2 controls
    '''
    # Just print out a board, and use string formatting 
    # along with conditional operators to display correct symbols
    #     Symbols:
    #       p1 - X
    #       p2 - O
    one = " "
    if 1 in p1:
        one = "X"
    elif 1 in p2:
        one = "O"
        
    two = " "
    if 2 in p1:
        two = "X"
    elif 2 in p2:
        two = "O"
        
    three = " "
    if 3 in p1:
        three = "X"
    elif 3 in p2:
        three = "O"
        
    four = " "
    if 4 in p1:
        four = "X"
    elif 4 in p2:
        four = "O"
        
    five = " "
    if 5 in p1:
        five = "X"
    elif 5 in p2:
        five = "O"
        
    six = " "
    if 6 in p1:
        six = "X"
    elif 6 in p2:
        six = "O"
        
    seven = " "
    if 7 in p1:
        seven = "X"
    elif 7 in p2:
        seven = "O"
        
    eight = " "
    if 8 in p1:
        eight = "X"
    elif 8 in p2:
        eight = "O"
        
    nine = " "
    if 9 in p1:
        nine = "X"
    elif 9 in p2:
        nine = "O"
        
    # Now print the board
    print"   _________________ "
    print"  |     |     |     | "
    print"  |  {}  |  {}  |  {}  | ".format(one, two, three)
    print"  |_____|_____|_____| "
    print"  |     |     |     | "
    print"  |  {}  |  {}  |  {}  | ".format(four, five, six)
    print"  |_____|_____|_____| "
    print"  |     |     |     | "
    print"  |  {}  |  {}  |  {}  | ".format(seven, eight, nine)
    print"  |_____|_____|_____| "
    print""
    
```

```python
def make_a_move():
    '''
        Function to get input from a user and verify its
        validity.
    '''
    # Get input from user
    move = raw_input("Enter your move: ")
    
    # Check for exit character
    if move == 'q' or move == "Q":
        return 'q'
    
    # Check to make sure input is legit
    try:
        move = int(move)
        return move
    except ValueError:
        print "You entered a non-integer. I can't work with that."
        return -1
```

```python
def check_for_winner(player):
    '''
        Function to check if someone has won the game. Checks if 
        a winning combination is held inside the array. The 
        winning combinations are hard coded, since there's a
        small number of them.
           Input parameters:
              player - list of squares controlled by player
        
    '''
    # Must be at least 3 entries to win
    if len(player) < 3:
        return False
    
    # List of sums (will be a victory if 
    # any of these has value 3 after loop)
    cols = [0]*3
    rows = [0]*3
    diag = [0]*2 # First entry is top left to bot right
                 # Second is top right to bot left
    
    # Check every square in players,
    # and add to appropriate sums
    for i in player:
        if i == 1:
            cols[0] += 1
            rows[0] += 1
            diag[0] += 1
        elif i == 2:
            cols[1] += 1
            rows[0] += 1
        elif i == 3:
            cols[2] += 1
            rows[0] += 1
            diag[1] += 1
        elif i == 4:
            cols[0] += 1
            rows[1] += 1   
        elif i == 5:
            cols[1] += 1
            rows[1] += 1
            diag[0] += 1
            diag[1] += 1
        elif i == 6:
            cols[2] += 1
            rows[1] += 1 
        elif i == 7:
            cols[0] += 1
            rows[2] += 1
            diag[1] += 1
        elif i == 8:
            cols[1] += 1
            rows[2] += 1 
        elif i == 9:
            cols[2] += 1
            rows[2] += 1
            diag[0] += 1
            
    # Check if any sum value is equal to 3
    # If there is, we have a winner
    if 3 in cols or 3 in rows or 3 in diag:
        return True
    else:
        return False
 
```

```python
def play_game():
    '''
       Function that starts and controls the whole tic-tac-toe
       game. 
    '''    
    
    # Game variables needed
    turn = 0   # Turn number
    used = []  # Squares already used
    player = [[],[]] # Player's moves
    no_winner = True
    no_quit = True
    
    # Welcome the user
    print " "*8 + "Welcome to Tic-Tac-Toe!" 
    print ""
    print "The following picture shows you how each square is labeled."
    print"   _________________ "
    print"  |     |     |     | "
    print"  |  1  |  2  |  3  | "
    print"  |_____|_____|_____| "
    print"  |     |     |     | "
    print"  |  4  |  5  |  6  | "
    print"  |_____|_____|_____| "
    print"  |     |     |     | "
    print"  |  7  |  8  |  9  | "
    print"  |_____|_____|_____| "
    print""
    print"To exit the game before it ends, simply enter q for a move."
    print""
    
    # Draw a blank board
    draw_board([],[])
    
    # Loop until game ends
    while turn < 9 and no_winner and no_quit:
               
        # Get the next move
        print "Player %d's Turn" % ((turn % 2) + 1)
        move = make_a_move()
        
        # Check if move is exit character
        if move == "q":
            no_quit = False
            
        # Make sure the square hasn't been used already
        while move in player[0] or move in player[1] or move not in range(10) and move != "q":
            print "That move has already been made or doesn't make sense. Choose another.\n"
            move = make_a_move()
 
        # Move is legit, add it to correct players squares
        if move != "q":
            player[turn % 2].append(move)
        
        # Update board
        draw_board(player[0], player[1])
               
        # Check for a winner
        if check_for_winner(player[turn % 2]):
            no_winner = False
            
        # Update turn count
        if no_winner:
            turn += 1
                 
    
    # Congratulate the winner
    if no_winner == False:
        print ""
        print "Player %d wins!" % ((turn % 2)+1)
    else:
        print ""
        print "Looks like we have a tie."
        
```

```python
play_game()
```
