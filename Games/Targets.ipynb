# Targets Game

This game takes a lot of code from the Molecular Dynamics 3 notebook. The object of this game is to fire a pellet from the center of the screen to hit a number of targets, all generated using Turtles. The player specifies the angle at which they want to fire the pellet, and has infinite tries.

PROBLEMS: So far this only works once, then you have to restart the kernel and run it again. Haven't figured out why this is.

```python
#import necessary packages:
import turtle
import random
import math

def play():
    random.seed()

    window = turtle.Screen()
    window.title('Pea-shooter Game')
    window.clear()

    # We will want to store these numbers relating to our window in a variable
    height = window.window_height()
    width = window.window_width()

    # Number of targets we want
    num_targets = 5

    # This initializes an empty list
    target_list = []

    # Now use a loop to initialize a new target
    # and do it num_targets times.
    for i in range(num_targets):
        target_list.append(turtle.Turtle())

    # Variables to hold things we want to be constant
    target_radius = 10
    
    # Scaling factor here is so we don't get our 
    # initial positions stuck on the edge.
    scaling_factor = 0.8
    
    # We need to loop over each target.
    for i in range(num_targets):
       
        # Draw the target in the proper shape, size, color, and
        target_list[i].shape('circle')
        target_list[i].shapesize(target_radius/10.0)
        target_list[i].color((random.random(),random.random(),random.random()))
        target_list[i].penup()
        target_list[i].goto(random.uniform(-1,1)*width/2.0 * scaling_factor, random.uniform(-1,1)*height/2.0 * scaling_factor)
      
        # Turtles can be very slow. This is a semi-fix to tell
        # turtles not to update the screen with every change,
        # but rather wait till a set of updates are done and 
        # then update the screen.
        #target_list[i].tracer(0,0)
        turtle.update()

    #make pellet
    pellet_radius = 5
    pellet = turtle.Turtle()
    pellet.shape('circle')
    pellet.shapesize(pellet_radius/10.0)
    pellet.color((0,0,0))
    pellet.penup()
    turtle.update()

    #can edit the pellet speed here
    pellet_speed = 10.0

    #this just initializes the angle theta. It must be initialized to something other than -1.0
    theta = 1
    
    #put the pellet in the center of the screen to start
    pellet.goto(0,0)
    
    #this will allow us to print the win statement to the screen only once
    printwin = True

    while theta != -1.0:
        #read in an angle from the user:
        theta = input("Please specify an angle between 0 and 360 (-1 to quit)")
    
        #because theta is a string, let's make it a float
        theta = float(theta)
    
        #give us a way out of the loop
        if theta == -1.0:
            break
    
        #use theta to set our pellet velocity
        x_velocity = pellet_speed*math.cos(math.radians(theta))
        y_velocity = pellet_speed*math.sin(math.radians(theta))

        #extract our pellet positions
        (x,y) = pellet.pos()

        #while loop will continue until the pellet is off-screen.
        while (-width/2.0 < x < width/2.0) and (-height/2.0 < y < height/2.0):
            #move the pellet forward a little in a straight line
            pellet.goto(x + x_velocity,y + y_velocity) #we can use dt = 1, so why even code it?
            
            #update position
            (x,y) = pellet.pos()
        
            #need to check for collision with a target
            collided = [] #a list of targets the pellet has collided with
            for i in range(len(target_list)): #loop over targets
                (u,v) = target_list[i].pos() #extract positions of target
                distance = math.sqrt((x-u)**2.0+(y-v)**2) #calculate distance from center of pellet to center of target
                if distance < target_radius + pellet_radius: #checks for collision
                    print("You hit one!")
                    collided.append(i) #record the index of the target that was hit
            collision_counter = 0; #count the number of collisions account for
            for i in collided: #loop over indices recorded
                
                target_list[i-collision_counter].color((1,1,1)) #change target to white
                
                #remove target from list of targets to be considered for future calculations
                target_list = target_list[:i-collision_counter] + target_list[i-collision_counter+1:] 
                
                collision_counter += 1
        
            #Now check to see if the player has won
            if len(target_list) == 0:
                if printwin:
                    print("You Win!")
                    printwin = False #Without this, the terminal will keep printing "You Win" until the pellet hits the wall
                theta = -1.0 #this is what terminates the outermost loop
        
            turtle.update()
    
        pellet.goto(0,0)
        turtle.update()

    del target_list
    window.bye()
```

```python
play()
```
